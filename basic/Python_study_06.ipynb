```python
# for in list

patterns = ['가위', '보', '보', '가위', '가위', '가위', '보', '가위', '바위', '보']

for i in patterns:
  print(i)
```

```python
# for in range
# 반복문을 이용해 원하는 횟수만큼 반복하는 방법
# 0, 1, 2, 3, 4를 순서대로 출력한다.

""" for i in [0,1,2,3,4]:
print(i)
"""
# 1부터 1000까지 출력하기
for i in range(1, 1001):
  print(i)
```

```python
# 교실에 4명의 학생이 있을 때 4명의 학생에게 순서대로 번호를 붙여서 이름과 함께 출력
# 1번: 뮤츠, 2번: 기라티나 3번: 파치리스 4번: 초염몽
names = ['뮤츠', '기라티나', '파치리스', '초염몽',]

for i in range(4):
  print('{}번: {}'.format(i+1, names[i]))
```

```python
# 교실에 4명의 학생이 있을 때 5명의 학생에게 순서대로 번호를 붙여서 이름과 함께 출력
# 1번: 뮤츠  2번: 기라티나  3번: 파치리스  4번: 초염몽  5번: 피카츄
names = ['뮤츠', '기라티나', '파치리스', '초염몽','피카츄']

for i in range(len(names)):
  print('{}번: {}'.format(i+1, names[i]))
```

```python
# 1번 뮤츠, 2번: 기라티나 3번: 파치리스 4번: 초염몽 5번: 피카츄
names = ['뮤츠', '기라티나', '파치리스', '초염몽','피카츄']

for i, name in enumerate(names):
  print('{}번: {}'.format(i+1, name))
```

```python
# range로 코드 작성해보기

print(chr(44032))
print(chr(44032+11172-1))
```

```python
# 한글 11172개 출력해보기

for i in range(11172):
  print('{}:{}'.format(i+44032, chr(44032+i)), end = '\t')
```

```python
# 자기이름 가장 빨리 출력해보기

print('내 이름은 {}{}{}입니다.'.format(chr(44608), chr(51116), chr(50896)))
```
